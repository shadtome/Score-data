```python
import importlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import base_line_models_ba_6 as blm
importlib.reload(blm)
```

```python
# Train Data
fd = '../../data/main_data/test/test_ba_6.csv'
test = pd.read_csv(fd)
test.head()
```

```python
# Train Data
fd = '../../data/main_data/train/train_ba_6.csv'
train = pd.read_csv(fd)
train.head()
```

```python
# Base Line linear model with all of the features:
linReg = blm.Simple_Linear_Regression(train)
linReg.perform_CV()
```

```python
players = ['bukayo saka','declan rice','jude bellingham','harry kane','cole palmer','martin degaard','cody gakpo',
           'gabriel jesus',]
for p in players:
    pred = linReg.predict_player(train,player=p)
    true = train[train['name'] == p].target_adjusted_market_value.values
    print(f'player "{p}" has prediction {pred} vs actual {true}\n')
```

```python
linReg.evaluate(test)
```
