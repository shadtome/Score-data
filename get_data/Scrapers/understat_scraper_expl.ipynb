```python
import requests
from bs4 import BeautifulSoup as bs
import os
import pandas as pd
import json
```

```python
url = 'https://understat.com/player/'
```

```python
page = requests.get(url + '/8497')
print(page.status_code)
```

```python
soup = bs(page.text,'html.parser')
```

```python
soup
```

```python
bad_script = soup.find_all('script')
print(bad_script)
string = bad_script[1].string
```

```python
ind_start = string.index("('")+2
ind_end = string.index("')")
json_data = string[ind_start:ind_end]
json_data = json_data.encode('utf8').decode('unicode_escape')

data = json.loads(json_data)
data = data['season']
data
```

```python
columns = ['position',
            'games',
            'goals',
            'shots',
            'time',
            'xG',
            'assists',
            'xA',
            'key_passes',
            'season',
            'team',
            'yellow',
            'red',
            'npg',
            'npxG',
            'xGChain',
            'xGBuildup']
column_data = dict()
for c in columns:
    column_data[c]=[]

```

```python
for d in data:
    for c in columns:
        print(type(d[c]))
        column_data[c].append(d[c])

print(column_data)
```

```python
df = pd.DataFrame(column_data)
df
```
