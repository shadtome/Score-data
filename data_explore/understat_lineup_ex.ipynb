```python
#Lets look at the most recent season games in the Premier League
con = sql.connect(path)

EPL_24 = pd.read_sql("""SELECT game.id,
                                game.season,
                                game.date,
                                game.team_h,
                                game.team_a,
                                game.league
                            FROM general_game_stats AS game
                            WHERE game.season='2024' AND game.league='EPL'
                                AND (game.team_h='Arsenal' OR game.team_a='Arsenal')
                            ;""",con)
con.close()
EPL_24
```

```python
con = sql.connect(path)

a_m_events = pd.read_sql("""SELECT events.minute,
                                    events.result,
                                    events.xG AS event_xG,
                                    events.player,
                                    events.situation,
                                    player.xG AS total_xG
                                FROM game_events AS events
                                JOIN lineup_stats AS player ON player.player_id = events.player_id
                                WHERE events.match_id = '26651' and player.match_id='26651'
                                ORDER BY events.minute;""",con)
con.close()
a_m_events
```

```python
display(lineup.columns)
con = sql.connect(path)

specific_lineup = pd.read_sql("""SELECT lineup.player,
                                        lineup.goals,
                                        lineup.shots,
                                        lineup.xG,
                                        lineup.position
                        FROM lineup_stats AS lineup
                        WHERE lineup.match_id = '26651';""",con)
con.close()
specific_lineup
```

```python
# I want to get the club stats with their ID, so it is easier to access in another table.
display(game_stats.columns)
con = sql.connect(path)

clubs= pd.read_sql("""SELECT club_id,
                            club,
                             league_id,
                            league,
                            season,
                            SUM(win) AS wins,
                            SUM(draw) AS draws,
                            SUM(loss) AS losses, 
                            SUM(win*3 + draw*1) AS pts,
                            AVG(xG) as avg_xG,
                            SUM(goals) AS total_goals,
                            SUM(goals_cond) AS total_goals_cond
                        FROM (
                                SELECT  h_id AS club_id,
                                        team_h AS club,
                                        league_id,
                                        league,
                                        season,
                                        h_goals AS goals,
                                        h_xg AS xG,
                                        CASE WHEN h_goals>a_goals THEN 1 ELSE 0 END AS win,
                                        CASE WHEN h_goals<a_goals THEN 1 ELSE 0 END AS loss,
                                        CASE WHEN h_goals = a_goals THEN 1 ELSE 0 END AS draw,
                                        a_goals AS goals_cond
                                FROM general_game_stats

                                UNION ALL

                                SELECT  a_id AS club_id,
                                        team_a AS club,
                                        league_id,
                                        league,
                                        season,
                                        a_goals AS goals,
                                        a_xg AS xG,
                                        CASE WHEN h_goals<a_goals THEN 1 ELSE 0 END AS win,
                                        CASE WHEN h_goals>a_goals THEN 1 ELSE 0 END AS loss,
                                        CASE WHEN h_goals = a_goals THEN 1 ELSE 0 END AS draw,
                                        h_goals AS goals_cond
                                FROM general_game_stats
                   
                                ) AS combined_teams
                            
                            GROUP BY season,club_id
                            
                            ORDER BY season;""",con)
con.close()

clubs.head(50)

#This table was integrated into the database as a easy way to access it
```
